```python
import pandas as pd
import tabula
```

```python
# extracting table from pdf using tabula

df = tabula.read_pdf('RMS-EKM-to upload.pdf',pages='101',lattice=True)

df
```

```python
df[0]
```

```python
# somehow the extraction contains a list of two duplicated dataframes of the table. Therefore selecting the first element of the list.

df = df[0]
```

```python
# the dataframe contains a lot of noise, for e.g., null columns with no headers. So dropping them.

df = df.dropna(axis=1,how='all')

df
```

```python
# the column header has shifted to the first row and sub-header to the second row. 

# edit: also it was noticed the final data contained '\r' in both columns and data. So removing it from the dataframe.

df = df.replace('\r',' ',regex=True)
```

```python
# Extracting the names of the places from first row (header row). Removing 'Trip','Day', and null value

lst_places = list(df.loc[0,:].dropna().unique())

lst_places
```

```python
for i in ['Trip','Day']:
    if i in lst_places:
        lst_places.remove(i)

lst_places
```

```python
# extracting only the body of the data as dataframe since the headers and sub-headers are messed up.
# we don't need 'Trip' column since we already have the index

df = df.iloc[2:,1:]

df
```

```python
# index is corrected so as to start from 0.

df.index-=2

df
```

```python
# rebuilding the headers and sub-headers using the list_places and MultiIndexing 
# added an empty element with 'Day' since multiindexing columns requires every header to have same no. of levels

lst_subheader = ['A','D','Via']

lst_columns = [('Day','')]

for i in lst_places:
    for j in lst_subheader:
        lst_columns.append((i,j))
        
lst_columns
```

```python
df_copy = df.copy()
df_copy.columns = pd.MultiIndex.from_tuples(lst_columns)

df_copy
```

```python
# we have created our dataframe.

# a mutliindexed dataframe with multiple column levels converted into a dictionary will have tuple keys.
# unfortunately dict with tuple keys do not convert well to JSON since we have to convert it back to dataframe
# therefore skip the multiindexing step and find a solution to building a dictionary apt to be converted to JSON
```

```python
# a dictionary will be created with 'Day'+places as keys and each place having a value:dict containing 'A','D','Via' as keys
# inturn 'A','D','Via' will have values:list(columnwise data)
```

```python
dict1 = {
    'Day':list(df.iloc[:,0])
}

dict1
```

```python
column_number=1
for i in lst_places:
    dict2={}
    for j in ['A','D','Via']:
        dict2[j]=list(df.iloc[:,column_number])
        column_number+=1
        dict1[i]=dict2

dict1
```

```python
# converting and exporting dict1 as JSON file

import json

with open("output.json", "w") as outfile: 
    json.dump(dict1, outfile)
```
